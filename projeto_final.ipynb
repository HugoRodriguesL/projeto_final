```python
from plyer import notification
import requests
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
```

# Conexão API

```python
url = ('https://brasilapi.com.br/api/ibge/uf/v1')
response = requests.get(url)
response
```

# Função alerta()

```python
def alerta(response):
    if response.status_code != 200:
        try:
            response.raise_for_status()  
        except Exception as erro:
            notification.notify(
                title=f'Erro: Status {response.status_code}',
                message=f'{erro}',  
                timeout=10
            )
```

# Tabelas (Data)

```python
if response.status_code == 200:
    data_json = response.json()
    print('Dados Recebidos!')
else:
    alerta(response)
```

```python
# tabela 1 - contendo sigla do Estado, nome e detalhamento sobre região.

df_estados = pd.DataFrame(data_json)
df_estados
```

```python
url = ('https://brasilapi.com.br/api/pix/v1/participants')
response = requests.get(url)
response
```

```python
if response.status_code == 200:
    data_json = response.json()
    print('Dados Recebidos!')
else:
    alerta(response)
```

```python
# Table 2 - contendo informações de transações realizadas via pix
df_pix = pd.DataFrame(data_json)
df_pix
```

```python
url = ('https://brasilapi.com.br/api/cvm/corretoras/v1')
response = requests.get(url)
response
```

```python
if response.status_code == 200:
    data_json = response.json()
    print('Dados Recebidos!')
else:
    alerta(response)
```

```python
# Table 3 - Dados de corretoras no BR
df_corretora = pd.DataFrame(data_json)
df_corretora
```

# Tratando os Dados

## df_estados

```python
df_estados.isnull().sum()
```

```python
# Agragando os dados da coluna ['regioes'] ao DF original.


df_regioes = pd.json_normalize(df_estados['regiao'])
df_regioes.columns = ['id_regiao', 'sigla_regiao', 'nome_regiao'] # Renomeia as colunas, evitando conflito/duplicidade.
df_estados_agg = pd.concat([df_estados, df_regioes], axis=1).drop('regiao', axis='columns').sort_values(by=['nome']) # Concatenando o df regioes ao DataFrame original (df_estados), e ordenando pelo nome do Estado.
```

```python

# Visualizando o resultado final
df_estados_agg
```

## df_pix

```python
df_pix.isnull().sum()
```

```python
# Tratamento de datas:

df_pix['inicio_operacao'] = pd.to_datetime(df_pix['inicio_operacao'])
df_pix['data_operacao'] = df_pix['inicio_operacao'].dt.date # Segregando a coluna com data
df_pix['hora_operacao'] = df_pix['inicio_operacao'].dt.strftime('%H:%M') # Segregando a coluna com hora e minuto
df_pix = df_pix.drop(columns=['inicio_operacao'])
```

```python

# DF_Pix final
df_pix
```

## df_corretora

```python
df_corretora.isnull().sum()
```

```python
# Conversao dos campos de data para datetime
df_corretora['data_patrimonio_liquido'] = pd.to_datetime(df_corretora['data_patrimonio_liquido'], errors='coerce')
df_corretora['data_inicio_situacao'] = pd.to_datetime(df_corretora['data_inicio_situacao'], errors='coerce')
df_corretora['data_registro'] = pd.to_datetime(df_corretora['data_registro'], errors='coerce')
df_corretora
```

# Conectando ao BD


```python
# Cria uma conexão com o banco de dados SQLite
engine = create_engine('sqlite:///projeto_final.db')

# Adiciona o DataFrame df_estados_agg à tabela 'estados' ao BD
df_estados_agg.to_sql('estados', con=engine, if_exists='replace', index=False)

# Adiciona o DataFrame df_pix à tabela 'pix' ao BD
df_pix.to_sql('pix', con=engine, if_exists='replace', index=False)

# Adiciona o DataFrame df_corretora à tabela 'corretora' ao BD
df_corretora.to_sql('corretora', con=engine, if_exists='replace', index=False)

print("Dados adicionados ao banco de dados com sucesso!")
```

```python
# Conectando ao banco
conn = sqlite3.connect('projeto_final.db')
cursor = conn.cursor()
```

```python
# Realizando leitura e retorno.
cursor.execute('''
SELECT * FROM estados WHERE sigla_regiao LIKE ('NE');
''')
leitura_exemplo = cursor.fetchall()
print(leitura_exemplo)
```

```python
conn.close() # Encerrando a conexão com o banco
```
