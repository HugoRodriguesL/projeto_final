```python
from plyer import notification
import requests
import pandas as pd
#from playsound import playsound
```

# Conexão API

```python
url = ('https://brasilapi.com.br/api/ibge/uf/v1')
response = requests.get(url)
response
```

# Função alerta()

```python
def alerta(response):
    if response.status_code != 200:
        try:
            response.raise_for_status()  
        except Exception as erro:
            notification.notify(
                title=f'Erro: Status {response.status_code}',
                message=f'{erro}',  
                timeout=10
            )
```

# Tabelas (Data)

```python
if response.status_code == 200:
    data_json = response.json()
    print('Dados Recebidos!')
else:
    alerta(response)
```

```python
# tabela 1 - contendo sigla do Estado, nome e detalhamento sobre região.

df_estados = pd.DataFrame(data_json)
df_estados
```
